## 傅里叶变换

### 图像中的高频和低频  
- 高频：变化剧烈地灰度分量，如边界
- 低频：变化缓慢的灰度分量，如大海

### 滤波器
- 低通滤波器：保留低频成分，会使图像变得更加模糊
- 高通滤波器：保留高频成分，会使图像变得更加清晰

### opencv中的滤波函数
- 主要使用cv2.dft()【傅里叶正变换】和cv2.idft()【傅里叶逆变换】，输入图像需要先转换成np.float32格式
- 得到的结果中频率为0的部分会在左上角，通常要转换到中心位置，可以通过shift变换来实现
- cv2.dft()返回值为双通道值(实部，虚部),通常还需要转换成图像格式才能展示

```python
# 关于dft的使用
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("shrink37.jpg", cv2.IMREAD_GRAYSCALE)  # 读取图片
img_float32 = np.float32(img)  # 将图像转换成float32格式，为dft变换的要求

dft = cv2.dft(img_float32, flags=cv2.DFT_COMPLEX_OUTPUT) # 设置为complex_output,输出与输入矩阵大小相同的矩阵
dft_shift = np.fft.fftshift(dft)
# print(dft_shift)
# print(dft_shift[:,:,0]) # 获得实部的矩阵
# print(dft_shift[:,:,1]) # 获得虚部的矩阵
# cv2.dft 变换后得到的结果是双通道的

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1])) # 对实部和虚部进行处理
# print(magnitude_spectrum)

plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.title('InputImage'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('magnitude_spectrum'), plt.xticks([]), plt.yticks([])
```

```python
# 实际应用与低通滤波以及高通滤波器
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("shrink37.jpg", cv2.IMREAD_GRAYSCALE)
img_float = np.float32(img)

dft = cv2.dft(img_float, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow, ccol = int(rows/2), int(cols/2)

# 低通滤波
# 制作掩码，与谱相乘，去除高频部分信息
mask = np.zeros((rows, cols, 2), np.uint8)
mask[crow-50:crow+50, ccol-50:ccol+50] = 1

idft = dft_shift*mask
idft_shift = np.fft.fftshift(idft)
img_back = cv2.idft(idft_shift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.title('Input Image'),plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(img_back, cmap='gray')
plt.title('Output'),plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# 实际应用与低通滤波以及高通滤波器
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("shrink37.jpg", cv2.IMREAD_GRAYSCALE)
img_float = np.float32(img)

dft = cv2.dft(img_float, flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

rows, cols = img.shape
crow, ccol = int(rows/2), int(cols/2)

# 高通滤波器
# 制作掩码，与谱相乘，去除高频部分信息
mask = np.ones((rows, cols, 2), np.uint8)
mask[crow-10:crow+10, ccol-10:ccol+10] = 0

idft = dft_shift*mask
idft_shift = np.fft.fftshift(idft)
img_back = cv2.idft(idft_shift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.title('Input Image'),plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(img_back, cmap='gray')
plt.title('Output'),plt.xticks([]), plt.yticks([])
plt.show()
```
