```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def CV2_show(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

模板匹配与卷积的原理很类似，通过将模板从原点开始滑动，逐个计算差别度，然后得到差别度最小的位置则为需要寻找的位置  
原图像大小为AxB，模板为axb，则输出为(A-a+1)x(B-b+1)

```python
img = cv2.imread("shrink37.jpg")
template = cv2.imread("shrink37_templet.jpg")
```

```python
h, w = template.shape[:2] # 获得前两个数
print(h, w)
```

```python
img.shape
```

```python
template.shape
```

- TM_SQDIFF 计算平方不同，计算值越小，越相关
- TM_CCORR 计算相关性，计算的值越大，越相关
- TM_CCOEFF 计算相关系数，计算出来的值越大，越相关
- 对归一化，越小则为越接近0，越大则为越接近1

```python
res = cv2.matchTemplate(img, template, cv2.TM_CCOEFF_NORMED)
res.shape
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
print(min_val, max_val, min_loc, max_loc)
```

```python
 methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCORR', 'cv2.TM_SQDIFF',
           'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF_NORMED']
    
for i in methods:
    img2 = img.copy()
    # 取出字符串的真值
    method = eval(i)
    # 进行模板匹配操作，得到结果
    res = cv2.matchTemplate(img2, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0]+w, top_left[1]+h) # loc为x,y的顺序
    
    # 画矩形
    cv2.rectangle(img2, top_left, bottom_right, (0, 0, 255), 1)
    
    plt.subplot(1,2,1), plt.imshow(res, cmap="gray")
    plt.xticks([]), plt.yticks([]) # 隐藏坐标轴
    plt.subplot(1,2,2), plt.imshow(img2, cmap="gray")
    plt.xticks([]), plt.yticks([])
    plt.suptitle(i)
    plt.show()
    
    
```

可以观察出来，对复杂图像的模板匹配方法实际上有一些误差且难以避免

```python
# 匹配多个图像
# 一般为使用归一化参数，然后取阈值

img2 = img.copy()
res = cv2.matchTemplate(img2, template, cv2.TM_CCORR_NORMED)
threshold = 0.99
loc = np.where(res >= threshold)
for pt in zip(*loc[::-1]):
    bottom_right = (pt[0]+w, pt[1]+h)
    cv2.rectangle(img2, pt, bottom_right, (0,0,255), 1)

CV2_show("img2", img2)
```
